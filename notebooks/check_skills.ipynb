# Check skills used in spell effects

The question:

**Which spells are missing a skill or have "unknown" as the skill?**

```python
from pathlib import Path

from opend6_tools.magic2 import *
from analyze_magic import all_books
```

```python
project_base = Path.home() / "Documents/Hobbies/Gaming/OpenD6"

books = all_books(project_base/"world")
```

```python
sum(len(b) for b in books.values())
```

```python
for name in books:
    for spell in books[name]:
        if spell.skill in {"unknown", None, ""}:
            print(name, spell.name, spell.skill)
```
